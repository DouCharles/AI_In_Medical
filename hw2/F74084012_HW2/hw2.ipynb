# Preprocess
classify and resize data

```python
import json
with open("nofind.json", "r") as file:
    nofind = json.load(file) 
with open("nodule.json", "r") as file:
    nodule = json.load(file) 
```

```python
from sklearn.model_selection import train_test_split

import os
import cv2

def generate_DataSet(size,dataPath,indexList):
    if not os.path.isdir(dataPath): os.makedirs(dataPath)
    for file, path in indexList:
        image = cv2.imread(path)
        image = cv2.resize(image,(size,size))
        cv2.imwrite(os.path.join(dataPath,file),image)

def train_valid_test_split(indexList,size):
    _, used_data = train_test_split(indexList, random_state=777, train_size=size)
    train_data, part_data = train_test_split(used_data, random_state=777, train_size=0.8)
    valid_data, test_data = train_test_split(part_data, random_state=777, train_size=0.5)
    return train_data, valid_data, test_data
```

```python
train_data, valid_data, test_data = train_valid_test_split(nofind,size=0.85)
generate_DataSet(size=256, dataPath="./dataset/trainset/nofind",indexList=train_data)
generate_DataSet(size=256, dataPath="./dataset/validset/nofind",indexList=valid_data)
generate_DataSet(size=256, dataPath="./dataset/testset/nofind", indexList=test_data)
```

```python
train_data, valid_data, test_data = train_valid_test_split(nodule,size=1)
generate_DataSet(size=256, dataPath="./dataset/trainset/nodule",indexList=train_data)
generate_DataSet(size=256, dataPath="./dataset/validset/nodule",indexList=valid_data)
generate_DataSet(size=256, dataPath="./dataset/testset/nodule",indexList=test_data)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
import os

import torch
from torch import nn
from torch import optim
import torch.nn.functional as F
from torchvision import datasets, transforms, models
from tqdm import tqdm
```

```python
train_dir = "/content/drive/MyDrive/aimas/dataset/trainset"
valid_dir = "/content/drive/MyDrive/aimas/dataset/validset"
```

```python
# Define transforms for the training data and testing data
train_transforms = transforms.Compose([
                    transforms.RandomRotation(30),
                    transforms.RandomResizedCrop(224),
                    transforms.RandomGrayscale(),
                    transforms.RandomHorizontalFlip(),
                    transforms.ToTensor(),
                    transforms.Normalize(
                        [0.485, 0.456, 0.406], 
                        [0.229, 0.224, 0.225])
                    ])


test_transforms = transforms.Compose([
                    # transforms.Resize(255),
                    transforms.CenterCrop(224),
                    transforms.ToTensor(),
                    transforms.Normalize(
                        [0.485, 0.456, 0.406],
                        [0.229, 0.224, 0.225]
                    )])


# Pass transforms in here, then run the next cell to see how the transforms look
train_data = datasets.ImageFolder(train_dir, transform=train_transforms)
valid_data = datasets.ImageFolder(valid_dir, transform=test_transforms)

trainloader = torch.utils.data.DataLoader(train_data, batch_size=96, shuffle=True)
validloader = torch.utils.data.DataLoader(valid_data, batch_size=64, shuffle=True)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
# model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
model = models.vgg11(weights=models.VGG11_Weights.IMAGENET1K_V1)
model.fc = nn.Sequential(
    nn.Linear(1000, 2, bias = True),
    nn.Sigmoid()
)
```

```python

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.000001)
lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size = 5, gamma = 0.5)

# Moves the model to either CPU or GPU
model.to(device)
```

```python
import numpy as np
import json
trainLossList = []
trainAccList = []

validLossList = []
validAccList = []


epochs = 25

minloss = np.Inf

for epoch in range(epochs):
    train_loss = 0.0
    train_total = 0
    train_correct = 0
    model.train()
    for images, labels in tqdm(trainloader):
        #Loading images and labels to device
        images = images.to(device)
        #labels = labels.type(torch.float32)
        labels = labels.type(torch.LongTensor)
        labels = labels.to(device)
        
        #Reseting Gradients
        optimizer.zero_grad()
        
        #Forward
        output = model(images)

        #Calculating Loss
        loss = criterion(output, labels)
        
        train_loss += loss.item()*images.size(0)
        
        pred = output.data.max(1, keepdim=True).indices
        ans = [int(labels[i]==pred[i]) for i in range(len(labels))]
        train_total += len(ans)
        train_correct += sum(ans)

        #Backward
        loss.backward()
        optimizer.step()
    trainLossList.append(train_loss/train_total)
    trainAccList.append(train_correct/train_total)
    torch.cuda.empty_cache()
    
    with torch.no_grad():
      model.eval()
      valid_loss = 0.0
      valid_total = 0  
      valid_correct = 0    
      for images, labels in tqdm(validloader):     
          #Loading images and labels to device
          images = images.to(device)
          # labels = labels.type(torch.float32)
          labels = labels.type(torch.LongTensor)
          labels = labels.to(device)
          
          #Reseting Gradients
          optimizer.zero_grad()
          
          #Forward
          output = model(images)

          #Calculating Loss
          loss = criterion(output, labels)
          
          valid_loss += loss.item()*images.size(0)
          
          pred = output.data.max(1, keepdim=True).indices
          ans = [int(labels[i]==pred[i]) for i in range(len(labels))]
          valid_total += len(ans)
          valid_correct += sum(ans)

    validLossList.append(valid_loss/valid_total)
    validAccList.append(valid_correct/valid_total)
                
    print(f"Epoch {epoch+1}/{epochs}.. ")
    print(f"Train loss: {train_loss/train_total}.. ")
    print(f"Test loss: {valid_loss/valid_total:.3f}.. ")
    print(f"Train accuracy: {(train_correct/train_total):.3f}")
    print(f"Test accuracy: {(valid_correct/valid_total):.3f}")
    if minloss > valid_loss/valid_total:
        minloss = valid_loss/valid_total
        print("Model Save")
        # torch.save(model.state_dict(), '/content/drive/MyDrive/aimas/DnC_ResNet50.pth')
        torch.save(model.state_dict(), '/content/drive/MyDrive/aimas/vgg/vgg11.pth')
    
    with open("/content/drive/MyDrive/aimas/vgg/train_loss.json", "w") as file:
        json.dump([float(v) for v in trainLossList],file) 
    with open("/content/drive/MyDrive/aimas/vgg/train_accuracy.json", "w") as file:
        json.dump([float(v) for v in trainAccList],file)
    with open("/content/drive/MyDrive/aimas/vgg/valid_loss.json", "w") as file:
        json.dump([float(v) for v in validLossList],file)
    with open("/content/drive/MyDrive/aimas/vgg/valid_accuracy.json", "w") as file:
        json.dump([float(v) for v in validAccList],file)
```

```python
## predict other data
import os
from PIL import Image 
from random import shuffle
from torchvision import datasets, transforms, models
import torch
from torch import nn
from tqdm import tqdm
testpath = '/content/drive/MyDrive/aimas/dataset/testset'


final_test_transforms = transforms.Compose([
                    transforms.CenterCrop(224),
                    transforms.ToTensor(),
                    transforms.Normalize(
                        [0.485, 0.456, 0.406], 
                        [0.229, 0.224, 0.225])
                    ])

final_test_data = datasets.ImageFolder(testpath, transform=final_test_transforms)




# test_data = torch.utils.data.DataLoader(testdatas, batch_size=1, shuffle=True)
test_data = torch.utils.data.DataLoader(final_test_data, batch_size=1, shuffle=True)
# model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
model = models.vgg11()
model.fc = nn.Sequential(
    nn.Linear(1000,2, bias = True),
    nn.Sigmoid()
)


model.load_state_dict(torch.load('/content/drive/MyDrive/aimas/vgg/vgg11.pth'))

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

model.to(device)
test_total = 0.0
test_correct = 0.0
with torch.no_grad():
  model.eval()
  for images, labels in tqdm(test_data):
    images = images.to(device)
    labels = labels.to(device=device, dtype=torch.int64)
    output = model(images)
    test_total += labels.size(0)
    _, predicted = torch.max(output.data,1)
    test_correct += (predicted == labels).sum().item()

print()
print(f"test accuraccy:{(test_correct/test_total):.3f}")
print("finish")
```
